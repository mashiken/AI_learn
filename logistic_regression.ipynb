```python
# ロジスティック回帰による分類器
```

```python
# ロジスティック回帰
# 通常の線形回帰をレベルアップした一般化線形回帰の1つ
# 分析の幅が広がる

# ロジスティック回帰の基本
# 0 から 1の値を返す 線形回帰式と同じ構造をロジット変換

# ロジスティック回帰の結果
# 0.5閾値 以上なら購入・以下なら不可
# 実際は確率値の分布をみて閾値を決め、適合率・再現率を見て閾値を判断する
# 適合率 購入すると予測して人の内本当に購入した人の割合
# 再現率 本当に購入した人の内、購入すると予測できた人の割合
# 適合率と再現率はトレードオフの関係にある為、バランスを見て閾値を決めるべき

# オッズ比 予測値の変化 オッズの比率の大きさが変数の影響度として評価
# オッズ ある事象が起こらない確率 ぶんの 起こる確率

# ロジスティック回帰分析の利用シーン
# 初期分析に使う事が多い 上司他の会社に説明をする事が多い
# 精度を求めるならXGBoost LightGBM 強力な機械学習手法
```

```python
import numpy as np
from sklearn import linear_model
```

```python
# 入力データの実例 2次元ベクトルと対応するラベルを定義
```

```python
x = np.array([[3.1,7.2],[4,6.7],[2.9,8],[5.1,4.5],[6,5],
              [5.6,5],[3.3,0.4],[3.9,0.9],[2.8,1],[0.5,3.4],
              [1,4],[0.6,4.9]])
y = np.array([0,0,0,1,1,1,2,2,2,3,3,3])
```

```python
# ラベル付きデータを用いて分類器を訓練する
# ロジスティック回帰による分類器LogisticRegressionの生成
```

```python
classifier = linear_model.LogisticRegression(solver='liblinear', C=2, multi_class='auto')
```

```python
classifier
```

```python
# 分類器を訓練
classifier.fit(x,y)
```

```python
%run utilities.ipynb
visualize_classifier(classifier, x, y)
```

```python
# C 値を100にすると境界がより正確になっている
classifier100 = linear_model.LogisticRegression(solver='liblinear', C=100, multi_class='auto')
classifier100.fit(x,y)
%run utilities.ipynb
visualize_classifier(classifier100, x, y)
```

```python
# C は分類間違いに対するペナルティとなるので、分類器は教師データにより適合する
# あまり大きな値を指定すると過剰適合になり、汎化性能が劣化するので注意
```

```python
classifier1000000 = linear_model.LogisticRegression(solver='liblinear', C=1000000, multi_class='auto')
classifier1000000.fit(x,y)
%run utilities.ipynb
visualize_classifier(classifier1000000, x, y)
```
